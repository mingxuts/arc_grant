```python
import pandas as pd
import numpy as np
import sys
import os
```

```python


# Name of the uni
# uq The University of Queensland
# um The University of Melbourne
# uts
# anu The Australian National University
# uwa The University of Western Australia
# usyd The University of Sydney
# unia The University of Adelaide
# monash Monash University
# unsw The University of New South Wales
PATH = "grants/"
os.listdir(PATH)
```

```python
def get_summary(uni_name):
    f_name = PATH + uni_name + "/output.csv"
    use_header = None
    df = pd.read_csv(f_name, header=use_header)
    df[4] = df[4].str.replace(",", "")
    col_names = ["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019"]
    arr_sum = [np.sum(pd.to_numeric(df[df[5] == i][4])) for i in range(2010, 2020)]
    r1 = pd.Series(arr_sum, index=col_names)
    proj_sum = [len(df.loc[df[5] == i]) for i in range(2010, 2020)]
    r2 = pd.Series(proj_sum, index=col_names)
    return r1, r2


r1, r2 = get_summary("uts")
cict = {"projs":r2, "amount": r1}
pd_um = pd.DataFrame(cict)
print(pd_um)
```

```python
unis = os.listdir(PATH)
# Number of Projs
noproj_dict = {}

# Amount of Fundings
amt_dict = {}

for u in unis:
    r1, r2 = get_summary(u)
    #key2 = u + 'proj'
    noproj_dict[u] = r2
    amt_dict[u] = r1

man_indx = ["anu", "monash", "um", "unia", "unsw", "uq", "usyd", "uwa", "uts"]    
pd_noproj = pd.DataFrame(noproj_dict)
pd_amt = pd.DataFrame(amt_dict)
print(pd_noproj[man_indx])
print(pd_amt[man_indx])
```

```python
sums = pd_noproj.select_dtypes(pd.np.number).sum().rename('total')
pd_noproj = pd_noproj.append(sums)
pd_noproj = pd_noproj[["anu", "monash", "um", "unia", "unsw", "uq", "usyd", "uwa", "uts"]]
pd_noproj
```

```python
man_indx = ["anu", "monash", "um", "unia", "unsw", "uq", "usyd", "uwa", "uts"]


sums = pd_amt.select_dtypes(pd.np.number).sum().rename('total')

    
pd_amt = pd_amt.append(sums)
pd_amt = pd_amt[man_indx]


for u in man_indx:
    pd_amt[u] = pd.to_numeric(pd_amt[u], downcast='float')
    

pd.options.display.float_format = '{:,.0f}'.format
pd_amt
```
